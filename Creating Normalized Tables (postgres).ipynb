```python
import psycopg2
```

```python
try:
    conn = psycopg2.connect("host=127.0.0.1 dbname=studentdb password=studentdb") 
except psycopg2.Error as e:
    print('Connection Error')
    print(e)

try:
    cur = conn.cursor()
except psycopg2.Error as e:
    print('Cursor Error')
    print(e)
conn.set_session(autocommit=True)   
```

```python
try:
    cur.execute("CREATE TABLE IF NOT EXISTS music_library (album_id int, \
                                                            album_name varchar, \
                                                            artist_name varchar, \
                                                            year int, \
                                                            songs text[]);")
except psycopg2.Error as e:
    print('Create Table Error')
    print(e)

try:
    cur.execute("INSERT INTO music_library (album_id, album_name, artist_name, year, songs) \
                VALUES (%s, %s, %s, %s, %s)", \
                (1, "Rubber Soul", "The Beatles", 1965, ["Michaelle", "Think For Yourself", "In my Life"]))
except psycopg2.Error as e:
    print('Insert Row Error')
    print(e)

try:
    cur.execute("INSERT INTO music_library (album_id, album_name, artist_name, year, songs) \
                VALUES (%s, %s, %s, %s, %s)", \
                (1, "Let it be", "The Beatles", 1970, ["Let it be", "Across the Universe"]))
except psycopg2.Error as e:
    print('Insert Row Error')
    print(e)
```

```python
try: 
    cur.execute("SELECT * FROM music_library;")
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)

row = cur.fetchone()
while row:
   print(row)
   row = cur.fetchone()
```
